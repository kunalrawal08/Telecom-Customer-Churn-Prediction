```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = dataframe = pd.read_csv(r"C:\Users\Kunal\OneDrive\Desktop\Telecom-Customer-Churn-Prediction\data\Telecom_customer churn.csv")
dataframe.head(5)
```

```python
# Summarize the dataset
print("Number of rows and columns:", dataframe.shape)
print("\nData types:")
print(dataframe.dtypes)

# Check for missing values
print("\nMissing values:")
print(dataframe.isnull().sum())

# Check class balance for churn
if 'churn' in dataframe.columns:
    print("\nClass balance:")
    print(dataframe['churn'].value_counts(normalize=True))
else:
    print("\n'Churn' column not found in the dataset.")
```

```python
# Simple cleaning of data
dataframe = dataframe.dropna()
```

WE NOW PERFORM EXPLAORATORY DATA ANALYSIS. INVOLVES PERFORMING IN-DEPTH UNIVARIATE ,BIVARIATE AND MULTIVARIATE ANALYSIS ON THE DATA.

```python
# Exploratory Data Analysis (EDA) using Matplotlib and Seaborn

# Visualize the distribution of churn
plt.figure(figsize=(10, 5))
sns.histplot(dataframe['churn'], kde=False)
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Frequency')
plt.show()
```

```python
# Univariate Analysis using Matplotlib and Seaborn

# Histogram for 'rev_Mean'
plt.figure(figsize=(7, 4))
sns.histplot(dataframe['rev_Mean'], kde=True)
plt.title('Histogram of rev_Mean')
plt.show()
```

```python
# Box plot for 'rev_Mean'
plt.figure(figsize=(7, 4))
sns.boxplot(y=dataframe['rev_Mean'])
plt.title('Box plot of rev_Mean')
plt.show()
```

```python
# Violin plot for 'rev_Mean'
plt.figure(figsize=(7, 4))
sns.violinplot(y=dataframe['rev_Mean'])
plt.title('Violin plot of rev_Mean')
plt.show()
```

```python
# Density plot for 'rev_Mean'
plt.figure(figsize=(7, 4))
sns.kdeplot(dataframe['rev_Mean'], shade=True)
plt.title('Density plot of rev_Mean')
plt.show()
```

```python
# Scatter plot for 'rev_Mean' against index
plt.figure(figsize=(7, 4))
plt.scatter(dataframe.index, dataframe['rev_Mean'])
plt.title('Scatter plot of rev_Mean against index')
plt.xlabel('Index')
plt.ylabel('rev_Mean')
plt.show()
```

```python
# Bivariate Analysis using Matplotlib and Seaborn

# Scatter plot for 'rev_Mean' vs 'churn'
plt.figure(figsize=(7, 4))
sns.scatterplot(x=dataframe['rev_Mean'], y=dataframe['churn'])
plt.title('Scatter plot of rev_Mean vs churn')
plt.show()
```

```python
# Hexbin plot for 'rev_Mean' vs 'churn'
plt.figure(figsize=(7, 4))
plt.hexbin(dataframe['rev_Mean'], dataframe['churn'], gridsize=50, cmap='Blues')
plt.colorbar(label='count')
plt.title('Hexbin plot of rev_Mean vs churn')
plt.show()
```

```python
# Box plot for 'rev_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.boxplot(x=dataframe['churn'], y=dataframe['rev_Mean'])
plt.title('Box plot of rev_Mean by churn')
plt.show()
```

```python
# Violin plot for 'rev_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.violinplot(x=dataframe['churn'], y=dataframe['rev_Mean'])
plt.title('Violin plot of rev_Mean by churn')
plt.show()
```

```python
# Bar plot for 'rev_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.barplot(x=dataframe['churn'], y=dataframe['rev_Mean'])
plt.title('Bar plot of rev_Mean by churn')
plt.show()
```

```python
# Bivariate Analysis using Matplotlib and Seaborn

# Scatter plot for 'mou_Mean' vs 'churn'
plt.figure(figsize=(7, 4))
sns.scatterplot(x=dataframe['mou_Mean'], y=dataframe['churn'], palette='viridis')
plt.title('Scatter plot of mou_Mean vs churn')
plt.show()
```

```python
# Hexbin plot for 'mou_Mean' vs 'churn'
plt.figure(figsize=(7, 4))
plt.hexbin(dataframe['mou_Mean'], dataframe['churn'], gridsize=50, cmap='plasma')
plt.colorbar(label='count')
plt.title('Hexbin plot of mou_Mean vs churn')
plt.show()
```

```python
# Box plot for 'mou_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.boxplot(x=dataframe['churn'], y=dataframe['mou_Mean'], palette='coolwarm')
plt.title('Box plot of mou_Mean by churn')
plt.show()
```

```python
# Violin plot for 'mou_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.violinplot(x=dataframe['churn'], y=dataframe['mou_Mean'], palette='Set2')
plt.title('Violin plot of mou_Mean by churn')
plt.show()
```

```python
# Bar plot for 'mou_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.barplot(x=dataframe['churn'], y=dataframe['mou_Mean'], palette='husl')
plt.title('Bar plot of mou_Mean by churn')
plt.show()
```

```python
# Line plot for 'mou_Mean' by 'churn'
plt.figure(figsize=(7, 4))
sns.lineplot(x=dataframe['churn'], y=dataframe['mou_Mean'], ci=None, marker='o', color='purple')
plt.title('Line plot of mou_Mean by churn')
plt.show()
```

```python
# Count plot for 'churn'
plt.figure(figsize=(7, 4))
sns.countplot(x=dataframe['churn'], palette='pastel')
plt.title('Count plot of churn')
plt.show()
```

```python
# Joint plot for 'mou_Mean' vs 'rev_Mean'
sns.jointplot(x='mou_Mean', y='rev_Mean', data=dataframe, kind='scatter', color='green')
plt.suptitle('Joint plot of mou_Mean vs rev_Mean', y=1.02)
plt.show()
```

```python
# Pair plot for selected features
sns.pairplot(dataframe, vars=['rev_Mean', 'mou_Mean', 'totmrc_Mean'], hue='churn', palette='coolwarm')
plt.suptitle('Pair plot of selected features', y=1.02)
plt.show()
```

```python
# Heatmap for a subset of features
plt.figure(figsize=(10, 8))
subset_corr = dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'churn']].corr()
sns.heatmap(subset_corr, annot=True, cmap='YlGnBu')
plt.title('Heatmap of selected features')
plt.show()
```

```python
# Multivariate Analysis using Seaborn's heatmap for a larger subset of features
plt.figure(figsize=(12, 10))
subset_corr = dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'ovrmou_Mean', 'churn']].corr()
sns.heatmap(subset_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of a larger subset of features')
plt.show()
```

```python
# Multivariate Analysis using Seaborn's heatmap for a larger subset of features
plt.figure(figsize=(12, 10))
subset_corr = dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'ovrmou_Mean', 'da_Mean', 'churn']].corr()
sns.heatmap(subset_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of a larger subset of features')
plt.show()
```

```python
# Multivariate Analysis using Seaborn's heatmap for another set of features
plt.figure(figsize=(12, 10))
subset_corr = dataframe[['ovrmou_Mean', 'da_Mean', 'vceovr_Mean', 'datovr_Mean', 'churn']].corr()
sns.heatmap(subset_corr, annot=True, cmap='viridis', fmt='.2f')
plt.title('Heatmap of another set of features')
plt.show()
```

```python
# Multivariate Analysis using 3D Scatter Plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dataframe['rev_Mean'], dataframe['mou_Mean'], dataframe['totmrc_Mean'], c=dataframe['churn'], cmap='coolwarm')
ax.set_xlabel('rev_Mean')
ax.set_ylabel('mou_Mean')
ax.set_zlabel('totmrc_Mean')
plt.title('3D Scatter Plot of rev_Mean, mou_Mean, and totmrc_Mean')
plt.show()
```

```python
# Multivariate Analysis using Parallel Coordinates
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(12, 6))
parallel_coordinates(dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'churn']], 'churn', colormap='coolwarm')
plt.title('Parallel Coordinates Plot')
plt.show()
```

```python
# Multivariate Analysis using Andrews Curves
from pandas.plotting import andrews_curves

plt.figure(figsize=(12, 6))
andrews_curves(dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'churn']], 'churn', colormap='viridis')
plt.title('Andrews Curves Plot')
plt.show()
```

```python
# Multivariate Analysis using RadViz
from pandas.plotting import radviz

plt.figure(figsize=(12, 6))
radviz(dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'churn']], 'churn', colormap='coolwarm')
plt.title('RadViz Plot')
plt.show()
```

```python
# Multivariate Analysis using PairGrid
sns.set(style="whitegrid")
g = sns.PairGrid(dataframe[['rev_Mean', 'mou_Mean', 'totmrc_Mean', 'churn']], hue="churn")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.suptitle('PairGrid of Selected Features', y=1.02)
plt.show()
```
